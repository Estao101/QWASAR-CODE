Import dataset from Kaggle

```python
import pandas as pd
```

Load dataset

```python
def load_dataset():
    df = pd.read_csv("googleplaystore.csv")
    return df
```

```python
dataset = load_dataset()
dataset
```

Summarize dataset

```python
def print_summarize_dataset(dataset):
    print("Dataset Summary:")
    print(dataset.info())
    print("\nFirst 5 rows of the dataset:")
    print(dataset.head())
    print("\nStatistical Summary:")
    print(dataset.describe())
```

```python
print_summarize_dataset(dataset)
```

Clean dataset

```python
def clean_dataset(dataset):
    dataset = dataset.dropna().drop_duplicates()
    dataset["Category"] = dataset["Category"].str.title()
    dataset["Installs"] = (dataset["Installs"].astype(str).str.replace(r'[^0-9]', '', regex=True).astype(int))
    return dataset
```

```python
cleaned_dataset = clean_dataset(dataset)
cleaned_dataset
```

```python
import matplotlib.pyplot as plt
```

```python
def print_histograms(dataset):
    #Plot with bar diagram the most populars paid apps of Family category
    family_paid = dataset[(dataset["Type"] != "Free") & (dataset["Category"] == "Family")]
    top_family_paid = family_paid.sort_values(by="Installs", ascending=False).head(10)
    plt.bar(top_family_paid["App"], top_family_paid["Installs"], color='skyblue')
    plt.xlabel("App")
    plt.ylabel("Number of Installs")
    plt.title("Top 10 Most Popular Paid Family Apps")
    plt.show()
    
```

```python
print_histograms(dataset)
```
